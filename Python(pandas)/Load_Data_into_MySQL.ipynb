```python
#Python Code:

import pandas as pd
import mysql.connector
```

```python
# MySQL database configuration
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': '3421',
    'database': 'retail',
}
```

```python
# CSV file path
csv_file_path = r"D:\Yess Info\Project\Retail Sales Performance Dashboard & Insights\data\superstore_cleaned.xlsx"

# ✅ Read CSV file and handle encoding
df = pd.read_excel(csv_file_path)

# ✅ Convert date columns properly (try both dayfirst & monthfirst)
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce', dayfirst=True, infer_datetime_format=True)
df['ship_date'] = pd.to_datetime(df['ship_date'], errors='coerce', dayfirst=True, infer_datetime_format=True)

# Check conversion result (optional)
print(df[['order_date', 'ship_date']].head())

# ✅ Connect to MySQL
cnx = mysql.connector.connect(**db_config)
cursor = cnx.cursor()

# ✅ Insert all rows safely
for i, row in df.iterrows():
    query = """INSERT INTO superstore_cleaned 
    (order_id, order_date, month, ship_date, customer, manufactory, product_name, segment, 
     category, subcategory, region, zip, city, state, country, discount, profit, quantity, sales, profit_margin)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"""

    order_date = row['order_date'].strftime('%Y-%m-%d') if isinstance(row['order_date'], pd.Timestamp) else None
    ship_date = row['ship_date'].strftime('%Y-%m-%d') if isinstance(row['ship_date'], pd.Timestamp) else None

    values = (
        str(row['order_id']),
        order_date,
        str(row['month']),
        ship_date,
        str(row['customer']),
        str(row['manufactory']),
        str(row['product_name']),
        str(row['segment']),
        str(row['category']),
        str(row['subcategory']),
        str(row['region']),
        str(row['zip']),
        str(row['city']),
        str(row['state']),
        str(row['country']),
        float(row['discount']),
        float(row['profit']),
        int(row['quantity']),
        float(row['sales']),
        float(row['profit_margin'])
    )

    cursor.execute(query, values)

# ✅ Commit after inserting all rows
cnx.commit()
print(f"✅ Inserted {cursor.rowcount} rows successfully!")

# ✅ Close connection
cursor.close()
cnx.close()
```
